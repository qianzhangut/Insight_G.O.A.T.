```python
df = pd.read_csv('C:/Users/Qian/python_work/webapp/static/data/cloud_data.csv')

background = np.array(Image.open("C:/Users/Qian/python_work/webapp/static/img/logo1.png"))

text = df.joint_tweets.loc[df.name =='Cristiano Ronaldo']
token = nltk.word_tokenize(text.values[0])
fdist = FreqDist(token)
tops=fdist.most_common(100)
text_cloud = [tops[i][0] for i in range(len(tops))]


text_plot = ' '.join([i for i in text_cloud])
lo = nltk.word_tokenize('Cristiano Ronaldo')
pl = [lo[0], lo[-1], lo[-1].lower(), lo[0].lower()]
stop = ['thi', 'come', 'don', 'citi', 'app', 'andi']
stopwords = stop + pl

wordcloud = WordCloud(width = 2000, height = 1500, 
            background_color ='white', margin=10, prefer_horizontal=1,
            stopwords = stopwords, mask = background,
            normalize_plurals=False, min_font_size = 20).generate(text_plot) 
# file_name= 'Erik Lamela'+'.png'
# file_path='C:/Users/Qian/python_work/webapp/static/img/' + file_name
# wordcloud.to_file(file_path)
```
