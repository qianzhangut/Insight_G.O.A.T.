```python
def get_user_info(twitter_user):
    """
    An example of using the query_user_info method
    :param twitter_user: the twitter user to capture user data
    :return: twitter_user_data: returns a dictionary of twitter user data
    """
    user_info = query_user_info(user= twitter_user)
    twitter_user_data = {}
    twitter_user_data["user"] = user_info.user
    twitter_user_data["fullname"] = user_info.full_name
    twitter_user_data["date_joined"] = user_info.date_joined
    twitter_user_data["id"] = user_info.id
    twitter_user_data["num_tweets"] = user_info.tweets
    twitter_user_data["following"] = user_info.following
    twitter_user_data["followers"] = user_info.followers
    twitter_user_data["likes"] = user_info.likes
    twitter_user_data["lists"] = user_info.lists
    
    return twitter_user_data
```

```python
ttt = []
for i in range(len(link_list)):
    try:
        ttt.append(get_user_info(link_list[i]))
    except:
        ttt.append({})
    
        
```

```python
for i in range(len(twit)):  
    ttt[i]['name'] = twit.loc[i, 'name'] 
```

```python
def mapp(feature):
    gh={}
    for i in range(len(ttt)):
        try:
            gh[ttt[i]['name']] = ttt[i][feature]
        except KeyError:
            gh[ttt[i]['name']] = 'NaN'
        df[feature] = df.name.apply(lambda x: gh[str(x)] if str(x) in gh else 0)     
    return df
```

```python
data_train['height_ver2'] = data_train['height'].apply(lambda x: str(x).replace(',', '').replace('m', ''))
```

```python
data_train['join_date_ver2'] = data_train['join_date'].apply(lambda x: (datetime.datetime.strptime("Dec31,2019", "%b%d,%Y").date() - datetime.datetime.strptime(x, "%b%d,%Y").date()).days if x!= '-' else 'NaN')
```

```python
data_train['exp_date_ver2'] = data_train['exp_date'].apply(lambda x: (datetime.datetime.strptime(x, "%d.%m.%Y").date() - datetime.datetime.strptime("Dec31,2019", "%b%d,%Y").date()).days if x!= '-' else 'NaN')
```

```python
data_train['date_joined_twitter_ver1'] = data_train['date_joined'].apply(lambda x: str(x).replace('.', '').replace('-', '').replace('AM', '').replace('PM', '').replace(' ', ''))
```

```python

data_train['date_joined_twitter_ver2'] = data_train['date_joined_twitter_ver1'].apply(lambda x: (datetime.datetime.strptime("Dec31,2019", "%b%d,%Y").date() - datetime.datetime.strptime(x, "%I:%M%d%b%Y").date()).days if x!= 'NaN' else 'NaN')
```

```python
def value_trans(value):
    if 'm' in value:
        a1 = value.replace('$','').replace('m', '') 
    elif 'k' in value:
        a1 = float(value.replace('$','').replace('k', ''))/1000
    return float(a1)

data_train['market_value_ver2'] = data_train['market_value'].apply(value_trans)
```
